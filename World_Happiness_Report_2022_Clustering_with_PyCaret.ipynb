```python
#import libraries
import pandas as pd
import numpy as np
from pycaret.clustering import *
```

```python
# load data
df=pd.read_csv("/content/World Happiness Report 2022.csv")
```

```python
df.head()
```

```python
setup(df) # setup the data
```

```python
kmeans=create_model('kmeans',num_clusters=5)
```

```python
plot_model(kmeans,plot='elbow') # use elbow method to see how many clusters, hence we need 5
```
